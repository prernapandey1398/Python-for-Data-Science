##### Encapsulation
+ It is one of the fundamental concept of OOPS.
+ It describes the idea of wrapping the data and a method that work on the data within one unit.
+ It is used to restrict access to method and variable.
+ Code and data are wrapped and make one single unit.

```python
class car:
    maxspeed=0
    name=''
    def __init__(self):
        self.maxspeed=250
        self.name="marcedise"
    def drive(self):
        print("driving with maximum speed"+ str(self.maxspeed))
    
r=car()
r.drive()
r.maxspeed=500
r.drive()
```

###### Data Abstraction


+ Data Abstraction and encapsulation are synonym.
+ It is used to hide internal details and so only required function.
+ Abstraction something means to give names to that thing. So, that the name capture basic idea of the function or whole program.
+ Only, showing the required things but not showing the internal functionality, how it works.
+ Data Abstraction = Data Encapsulation + Data Hiding

```python
#Data Abstraction
class roboot:
    def __init__(self, name=None):
        self.name=name
    def say_hi(self):
        if self.name:
            print("Hello !!"+ self.name)
        else:
            print("Hii !!")
            
    def set_name(self, name):
        self.name=name
    def get_name(self):
        return self.name
x=roboot()
x.set_name("P2")
x.say_hi()
y=roboot()
y.set_name(x.get_name())
print(y.get_name())
```

##### Multithreading
+ It is an entity  with a process can be scheduled for execution.
+ It is the smallest unit of a process that can be performed in an OS.
+ In simple word a thread is a sequence of instruction within a program that can be executed independently of other code.
+ A thread contains all that information in the thread control block.

+ Multi thread can exist within one process where

  + 2. All threads of the process share global variable and the program code.

```python
# multithreading
import threading
def print_cube(num):
    """
    function to print cube of given number
    """
    print("Cube: {}".format(num*num*num))
def print_square(num):
    """
    function to print square of given number
    """
    print("Square: {}".format(num*num))
    
if __name__=="__main__":
    t1=threading.Thread(target=print_square, args=(10,))
    t2=threading.Thread(target=print_cube, args=(10,))
    
    t1.start()
    t2.start()
    
    #wait untill the thread complete 
    t1.join()
    t1.getName()
    t1.setName("Prerna")
    t1.isDaemon()
    t1.is_alive()
    t2.join()
    
    print("Done")
```

##### Method Overloading

+ Method overloading is nothing but a polymorphism at the time of compile our code.
+ In this more than one method of the class share the same method name having different signature.
+ Method overloading is used to add more behaviour method.

```python
def add(a,b):
    return a+b
def add(a,b,c):
    return a+b+c
add(2,3)
```

```python
# First product method.
# Takes two argument and print their
# product
def product(a, b):
    p = a * b
    print(p)
     
# Second product method
# Takes three argument and print their
# product
def product(a, b, c):
    p = a * b*c
    print(p)
# Uncommenting the below line shows an error    
# product(4, 5) gives error
 
# This line will call the second product method
product(4, 5, 5)
```

##### Method overriding

+ Method overriding is an example of a runtime polymorphism.
+ In this specific implementation of the method that is already provided by parent class to the child class.
+ Inheritance always required between parent and child in the method overriding.

```python
class A:
    def sayhi(self):
        print("I am in class A")
class B(A):
    def sayhi(self):
        print("I am in class B")
b= B()
b.sayhi()
```

```python
class A:
    def fun1(self):
        print('feature_1 of class A')

    def fun2(self):
        print('feature_2 of class A')
# Modified function that is 
# already exist in class A
class B(A):
    def fun1(self):
        print('Modified feature_1 of class A by class B') 
    def fun3(self):
        print('feature_3 of class B')

# Create instance
obj = B()
# Call the override function
obj.fun1()
```
